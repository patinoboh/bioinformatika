```python
df = pd.read_csv(data_path, sep="\t", comment="!", header=0)
print(df.head())
```

```python
# transpose the dataframe so that the samples are in rows and the genes are in columns
df = df.transpose()
# print(df.head())
```

```python
times_row = "!Sample_title"
submission_date_row = "!Sample_submission_date"
contact_name_row = "!Sample_contact_name"
data_processing_row = "!Sample_data_processing"
scan_protocol_row = "!Sample_scan_protocol"
hyb_protocol_row = "!Sample_hyb_protocol"
extract_protocol_row = "!Sample_extract_protocol_ch1"
molecule_row = "!Sample_molecule_ch1"
sex_row = "!Sample_characteristics_ch1" # first
gestational_age_row = "!Sample_characteristics_ch1" # second
smoke_type_row = "!Sample_characteristics_ch1" # third
debug = "\"cg00000029\""

wanted_rows = [
    times_row,
    submission_date_row,
    contact_name_row,
    data_processing_row,
    scan_protocol_row,
    hyb_protocol_row,
    extract_protocol_row,
    molecule_row,
    sex_row,
    gestational_age_row,
    smoke_type_row,
    debug
]

# open only first 100 lines of the file to find the wanted rows

with open(data_path, "r") as file:
    for i, line in enumerate(file):
        if i >= 100:
            break
        if line.startswith(tuple(wanted_rows)):
            data_row = line.strip().split("\t")
            # add the data_row to the data dataframe
            df_tmp = pd.DataFrame([data_row], columns=df.columns)
            df = pd.concat([df, df_tmp], ignore_index=True)


# print(df.tail(12))
```
